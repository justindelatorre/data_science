#$k$-NN Classification

**$k$-Nearest Neighbors** ($k$-NN) is a supervised machine learning classification algorithm that uses a data point's most similar "neighbors" to help label it. The $k$ represents an optimal number of neighbors that will help identify the class of a non-labelled data point.

To review, a **supervised** machine learning algorithm is one in which a portion of pre-labelled data is used to "train" the algorithm. Then, the algorithm is tested with the remainder of the data. If the test returns satisfactory results (i.e., a certain level of acceptable accuracy), then it the algorithm is used on new, unlabelled data.

A **classification** algorithm is used on categorical data, in which the data are discrete, rather than continuous.

Let's start by loading all the necessary packages and libraries, including the data from the [Fisher Iris data set](http://en.wikipedia.org/wiki/Iris_flower_data_set), which are the data we'll be exploring.

```python
%matplotlib inline
from __future__ import division
import pandas as pd
import numpy as np
from seaborn import plt
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets, feature_selection
from sklearn.cross_validation import train_test_split, cross_val_score

# Load in the data
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Currently, you'll see that the iris_df DataFrame contains no labels, so let's add them from iris.target.
iris_df['Target'] = iris.target

print iris_df.head()
```

Let's take a look at the options we have for the plot() method and plot the iris data.

```python
iris_df.plot?
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3)

colors = ['r','g','b']
for i in range(3): 
    tmp = iris_df[iris_df.Target == i]
    tmp.plot(x=0,y=1, kind='scatter', c=colors[i], ax=axes[0,0])

for i in range(3): 
    tmp = iris_df[iris_df.Target == i]
    tmp.plot(x=0,y=2, kind='scatter', c=colors[i], ax=axes[0,1])

for i in range(3): 
    tmp = iris_df[iris_df.Target == i]
    tmp.plot(x=0,y=3, kind='scatter', c=colors[i], ax=axes[0,2])
    
for i in range(3): 
    tmp = iris_df[iris_df.Target == i]
    tmp.plot(x=1,y=2, kind='scatter', c=colors[i], ax=axes[1,0])

for i in range(3): 
    tmp = iris_df[iris_df.Target == i]
    tmp.plot(x=1,y=3, kind='scatter', c=colors[i], ax=axes[1,1])

for i in range(3): 
    tmp = iris_df[iris_df.Target == i]
    tmp.plot(x=2,y=3, kind='scatter', c=colors[i], ax=axes[1,2])
```

In the scatter plots above, we're essentially plotting features of the dataset against each other in order to develop some intuition of how the data are potentially correlated.

Now we're going to use scikit-learn's **<code>test_train_split()</code>** function to create the training and test data sets. This is a common workflow step used to prep data for use with many algorithms, not only $k$-NN.

Four variables are produced: two training sets and two test sets. The first parameter is the features that are being evaluated; the second are the labels; the third parameter indicates the proportion of the total data to be used as test data; and the fourth (<code>random_state</code>) is a pseudo-random number generator to help with sampling.

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=12)
```

Before we move on, let's create values of $k$ that we'll test on the iris data.

```python
n_neighbors = range(1, 51, 2)
print n_neighbors
```

Now, let's loop through these values and test to see how well each one fares when we use the **<code>KNeighborsClassifier()</code>** method.

```python
scores = []
for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train, y_train) # fit the model using the training data
    scores.append(clf.score(X_test, y_test)) # score each k-value using the test data, then append to the list
```

Let's plot each $k$-value score.

```python
plt.plot(n_neighbors, scores, linewidth=3.0)
```

###Some Things to Consider
<i>Why does the classification rate go down with more neighbors?</i><br><br>
It seems that, past ~11 $k$, the model overfits the data.

<i>If we have N points in our dataset, what would happen if we use N neighbors to classify each point?</i><br><br>
If $k$ == N, then the model would severely overfit the data, and each data point's "neighbor" would only be itself.

##Cross-Validation

**Cross-validation** is a method that allows us to confirm the accuracy of a model by running the same test/train/split workflow on different partitions of the data, rather than only one one partition combination.

<img src="cv_diagram.png">

Let's reload the iris data again, but this time, let's import **<code>cross_val_score()</code>** from scikit-learn's <code>cross_validation</code> package as well and score the model accuracy using that with $k$ = 11 and setting the number of folds (<code>cv</code>) to 5, which is equivalent to <code>test_size=0.2</code> in <code>test_train_split()</code>.

```python
from sklearn.datasets import load_iris
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

from sklearn.cross_validation import cross_val_score #import cross_val_score 

clf = neighbors.KNeighborsClassifier(11, weights='uniform')
clf.fit(iris.data, iris.target)
scores = cross_val_score(clf, iris_df.values, iris.target, cv=5)
```

Let's print the scores for each permutation, then calculate the mean of the scores to get the overall accuracy.

```python
print scores
print scores.mean()
```

Based on the values generated by cross-validation above, $k$ = 11 returns an average 98% accuracy.
